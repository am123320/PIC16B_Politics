```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("final.csv")
all_cols = df.columns.tolist()
print(all_cols)
```

```python
df_ca = df[df['State'] == 'California']

race_columns = ['White', 'White (non Hispanic or Latino)', 'Hispanic or Latino', 
                'Black', 'American Indian', 'Asian', 'Pacific', 'Multiracial']
race_data = df_ca[race_columns]

plt.figure(figsize=(12, 8))
race_data.plot(kind='bar', stacked=True, colormap='viridis', alpha=0.7)
plt.title('Race and Ethnicity Distribution')
plt.xlabel('District')
plt.ylabel('Population')
plt.legend(title='Race/Ethnicity')
plt.show()
```

```python
race_cols = ['Year', 'State', 'District',
             'Democratic Votes', 'Republican Votes', 'Total Votes', 'Democratic Percentage','Republican Percentage',
             'Voting Eligible Male', 'Voting Eligible Female', 'Voting Eligible Population', 'Total Population', 'Voter Turnout',
             'White', 'White (non Hispanic or Latino)', 'Hispanic or Latino', 'Black', 'American Indian', 'Asian', 'Pacific', 'Multiracial']

df_race = df[race_cols]
```

```python
df_ca_race = df_race[df_race['State'] == 'California']

race_cols = ['White', 'Black', 'American Indian', 'Asian', 'Pacific', 'Multiracial']
df_ca_race = df_ca_race.copy()
df_ca_race['Race'] = df_ca_race[race_cols].idxmax(axis=1)

melted_df = pd.melt(df_ca_race, id_vars=['Race'], value_vars=['Democratic Percentage', 'Republican Percentage'])

plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Race', y='value', hue='variable', data=melted_df, palette={'Democratic Percentage': 'blue', 'Republican Percentage': 'red'})
bar_plot.set(title='Democratic and Republican Percentages by Race in California')
bar_plot.set_ylabel('Percentage')
plt.show()
```
